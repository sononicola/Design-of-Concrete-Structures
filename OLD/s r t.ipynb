### Flessione retta in semplice armatura

```python
sigmasr = 360
n = 15

sigmac=np.arange(3,24.5,.5)
s = n * sigmac / (n*sigmac + sigmasr)
r = sqrt(2/(sigmac * s * (1-s/3)))
t = 1 / (sigmasr * r * (1-s/3))
tab=table(columns=[sigmac,s.round(3),r.round(3),t.round(6)], header_row=[r"$\sigma_c$",r"$s$",r"$r$",r"$t$"], align='right')
tab
```

### Flessione retta in doppia armatura

```python
beta = 0.5 # beta = As'/ As
delta2 = 0.10 # delta'' = d'' / d

sigmasr = 360
n = 15

sigmac = np.arange(3,24.5,.5)
s = n * sigmac / (n*sigmac + sigmasr)
den = 1 + (s-delta2)/(1-s) * (beta*(s/3 - delta2))/(1 - s/3)
r1 = sqrt( 2/(sigmac * s * (1 - s/3)) )  *  sqrt( (1 - beta * ((s-delta2)/(1-s))) / den )
t1 = 1 / (sigmasr * r1 * (1-s/3) * den) 
tab = table(columns=[sigmac,s.round(4),r1.round(4),t1.round(7)], header_row=[r"$\sigma_c$",r"$s$",r"$r'$",r"$t'$"], align='right')
tab
```

```python
latex(tab)
```
